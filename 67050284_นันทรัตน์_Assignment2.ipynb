```python
import numpy as np
import pandas as pd
import scipy.stats as stats
df = pd.read_csv('BloodPressure.csv')
```

```python
df[['bp_before','bp_after']].describe()
```

```python
df.loc[df["agegrp"]=="46-59"] [['bp_before','bp_after']].describe()
```

```python
df.loc[df["agegrp"]=="60+"] [['bp_before','bp_after']].describe()
```

```python
significance = 0.01
```

```python
stat,p_value = stats.ttest_rel(df['bp_before'].loc[df["agegrp"]=="30-45"], df['bp_after'].loc[df["agegrp"]=="30-45"])
```

```python
print('t_stat: %0.4f, p_value: %0.4f' % (stat, p_value))
if p_value < significance:
  print ("Reject the null hypothesis")
else:
  print ("Accpet the null hypothesis")
```
